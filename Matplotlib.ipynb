```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,100,100)
y1 = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)
```

```python
plt.plot(x,y1, label='sin')
plt.plot(x,y2, label='cos')
plt.xlim(x.min()*0.5, x.max()*1.2)
plt.ylim(y1.min()*1.1, y1.max()*1.5)
plt.legend()
plt.savefig('./matplotlib_img/line_plot1.png', dpi=300)
```

```python
df = pd.DataFrame(columns=['Date','Population'])
df['Date'] = pd.date_range(start='01/01/2019', periods=18, freq='m')
df['Population'] = np.random.randint(150,1500,18)
df.plot.line(x='Date', y='Population', color='green')
plt.legend()
plt.savefig('./matplotlib_img/line_plot2.png', dpi=300)
```

```python
df.set_index('Date', drop='Date', inplace=True)
df['Total_Population'] = df['Population'].cumsum()
df.plot.area()
plt.savefig('./matplotlib_img/area_plot1.png', dpi=300)
```

```python
x = np.linspace(0,10,50)
y1 = x **2
y2 = x[::-1] **2
colors = np.random.rand(50)
area = (30 * np.random.rand(50))**2
plt.scatter(x,y1, s=200, marker='*', color='lightgreen', label='y1')
plt.scatter(x,y2, s=area, c=colors, alpha=0.5, label='y2')
plt.legend()
plt.savefig('./matplotlib_img/scatter_plot1.png', dpi=300)
```

```python
df = pd.DataFrame(columns=['x','y1','y2'])
df['x'], df['y1'], df['y2'] = x, y1+np.random.randint(-10,10,50), y2+np.random.randint(-5,15,50)
fig = df.plot.scatter(x='x', y='y1',label='y1' )
fig.scatter(x=df['x'], y=df['y2'],color='green')
fig.legend()
plt.savefig('./matplotlib_img/scatter_plot2.png', dpi=300)
```

```python
plt.plot(x,y1, color='green')
plt.scatter(x,y2, color='blue', marker='^')
plt.xlabel('Angle')
plt.ylabel('Value')
plt.savefig('./matplotlib_img/line_scatter_plot1.png', dpi=300)
```

```python
a1 = np.random.normal(loc=12, scale=10, size=150)
a2 = np.random.normal(loc=2, scale=10, size=150)
fig = plt.hist([a1, a2],25, alpha=0.5 ,  label=['a1', 'a2'])
plt.legend()
plt.savefig('./matplotlib_img/hist1.png', dpi=300)
```

```python
df = pd.DataFrame(columns=['a1', 'a2'])
df['a1'], df['a2'] = a1, a2
df.plot.hist(bins=12, alpha=0.5)
plt.savefig('./matplotlib_img/hist2.png', dpi=300)
```

```python
df.plot.kde()
plt.savefig('./matplotlib_img/hist3.png', dpi=300)
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.bar(rot=0, figsize=(8, 5))
plt.savefig('./matplotlib_img/bar_plot1.png', dpi=300)
```

```python
df.plot.pie(y='speed', figsize=(10, 8))
plt.savefig('./matplotlib_img/pie_chart1.png', dpi=300)
```

```python
fig = plt.pie(speed, labels=index, autopct='%1.1f%%')
plt.savefig('./matplotlib_img/pie_chart2.png', dpi=300)
```

```python
fig = df.plot.pie(subplots=True, figsize=(18, 8))
plt.savefig('./matplotlib_img/pie_chart3.png', dpi=300)
```
